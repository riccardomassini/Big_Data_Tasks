**Task2a: Implementare in Python l’esempio della somma di due dadi.
Usando la libreria matplotlib, crearne poi il grafico pdf**

Questo script genera un grafico che mostra con quale probabilità una certa somma di due valori estratti dopo un lancio di due dadi a 6 facce occorre. Per fare ciò, vengono generate tutte le possibili combinazioni di due lanci tramite la funzione product della libreria itertools, considerando che ogni dado può assumere valori da 1 a 6. Successivamente, si calcola la somma per ogni coppia di valori ottenuta e utilizzando numpy, si determinano i valori unici delle somme e la loro frequenza di occorrenza. Il grafico risultante, realizzato con matplotlib, utilizza un diagramma a gradini (step plot) per rappresentare la distribuzione di probabilità della somma dei due dadi. È possibile osservare come la distribuzione segua un andamento triangolare: le somme più basse (2 e 3) e più alte (11 e 12) hanno probabilità inferiori, mentre quelle centrali (come 7) sono le più probabili, in quanto possono essere ottenute in più modi diversi.

````python
import matplotlib.pyplot as plt
from itertools import product
import numpy as np

# Calcolo di tutte le possibili combinazioni e le relative somme
combinations = list(product(range(1, 7), repeat=2)) # Genera le possibili coppie
sums = [sum(pair) for pair in combinations] # Somme per ogni combinazione

# Calcolo della frequenza (quante volte una somma appare), valori unici (somme distinte) e della probabilità
unique_values, frequencies = np.unique(sums, return_counts=True)
probabilities = frequencies / sum(frequencies)

# Creazione del grafico step delle probabilità (grafico a gradini)
plt.step(unique_values, probabilities, where='mid', color='red', linewidth=2)
plt.title('Distribuzione della somma di due dadi (pdf)', fontsize=10)
plt.xlabel('Somma dei due dadi', fontsize=10)
plt.ylabel('Probabilità', fontsize=10)
plt.xticks(range(2, 13)) # Mostra i valori sull'asse delle x per le somme che ci servono
plt.grid(axis='y', linestyle='-', alpha=0.7)

# Mostra il grafico
plt.show()
````

**TASK2b: A partire dai dataset visti per l’es. ProduttoreA e
ProduttoreB di smartphone calcolare Media e Dev. Standard (come da esempio Excel) usando la libreria Python numpy e graficare la distribuzione dei dati usando la libreria Python matplotlib**

Questo script riutilizza i dati analizzati nella task02 dove venivano descritte le distribuzione normali e relative funzioni di densità di due produttori di smartphone.
Vengono calcolate la media e la deviazione standard dei valori in esame tramite i metodi "mean" e "std" presenti nella libreria numpy.
Per la visualizzazione della distribuzione era prima necessario definire la funzione di densità mediante il metodo "stats.norm.pdf" della libreria "scipy.stats" che prende come parametri i dati in oggetto, la media e la deviazione stardard di tali valori.
Infine grazie alla libreria "matplotlib.pyplot", vengono creati e poi mostrati i grafici di queste due distribuzioni.
Si può notare una differenza sostanziale tra i grafici delle due funzioni, infatti il produttore A mostra una curva a campana ampia e questo indica che i valori sono distribuiti su un ampio intervallo, il valore tipico di produzione è ben distribuito, con alcuni prodotti migliori e altri peggiori. Essendoci anche una deviazione standard alta, poichè la curva è larga, mostra una grande differenza tra i prodotti, alcuni molto buoni e altri di qualità inferiore. Al contrario, il produttore B presenta una distribuzione molto più concentrata attorno a un unico valore, con una curva stretta e una deviazione standard molto bassa. Questo indica che i prodotti di questo produttore hanno una qualità più uniforme e costante, con poche variazioni tra un'unità e l'altra.

````python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Definisce l'URL pubblico del file excel
public_url = f"https://docs.google.com/spreadsheets/d/{id_2}"

# Legge un file excel che viene scaricato tramite l'endpoint /export
df = pd.read_excel(public_url + "/export?format=xlsx")

# Estrae le colonne dal DataFrame e le converte in liste
producerA_data = df["Vita Batteria A"].to_list()
producerB_data = df["Vita Batteria B"].to_list()

# Calcolo di media e deviazione standard
mean_A = np.mean(producerA_data)
std_dev_A = np.std(producerA_data)
mean_B = np.mean(producerB_data)
std_dev_B = np.std(producerB_data)

# Calcolo delle funzioni di densità di probabilità
pdf_A = stats.norm.pdf(producerA_data, loc=mean_A, scale=std_dev_A)
pdf_B = stats.norm.pdf(producerB_data, loc=mean_B, scale=std_dev_B)

_, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 4)) # Tupla per rappresentare i 2 grafici

# Creazione del grafico per il produttore A
ax1.plot(producerA_data, pdf_A, label="Distribuzione Normale Produttore A", color='blue')
ax1.set_title("Distribuzione Normale Produttore A")
ax1.set_xlabel("Valore")
ax1.set_ylabel("Densità di probabilità")
ax1.legend()
ax1.grid(True, linestyle="--", alpha=0.6)
ax1.set_xlim(0, 100)

# Creazione del grafico per il produttore B
ax2.plot(producerB_data, pdf_B, label="Distribuzione Normale Produttore B", color='red')
ax2.set_xlabel("Valore")
ax2.set_ylabel("Densità di probabilità")
ax2.set_title("Distribuzione Normale Produttore B")
ax2.legend()
ax2.grid(True, linestyle="--", alpha=0.6)
ax2.set_xlim(0, 100)

# Mostra il grafico
plt.show()
````

# **Correlazione**

**TASK3a: Implementare la funzione valore assoluto 𝑦 = |𝑥| usando sia Excel che Python e calcolarne il coeff. di correlazione di Pearson**

In questa task è stata rappresentata la funzione valore assoluto y = |x| per i valori interi che vanno da -4 a 4 sia in excel che in python. L'indice di correlazione lineare in seguito calcolato e con risultato 0, indica che non esiste una relazione lineare tra le variabili, ma potrebbero esserci relazioni di altro tipo.

Il coefficiente di Pearson misura il gardo di relazione lineare tra 2 variabili (varia tra -1 e 1).

image.png

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Creazione array di valori e calcolo del valore assoluto di ciascun elemento
x = np.arange(-4, 5)
y = np.abs(x)

# Creazione tabella (valore originale - valore assoluto)
df = pd.DataFrame({"Valore originale": x, "Valore assoluto": y})

# Creazione del grafico a punti (scatter plot) con etichette e titolo
plt.figure(figsize=(6, 4))
plt.scatter(x, y, color='r', label="y = |x|", marker='o')
plt.xlabel("Valore originale (x)")
plt.ylabel("Valore assoluto (|x|)")
plt.title("Grafico a punti della funzione valore assoluto")
plt.axhline(0, color='gray', linewidth=0.5)  # Linea orizzontale su y=0
plt.axvline(0, color='gray', linewidth=0.5)  # Linea verticale su x=0
plt.grid(True, linestyle="--", alpha=0.7)
plt.legend()

# Mostra il grafico
plt.show()

# Calcolo e stampa della correlazione di Pearson
pearson = df.corr().iloc[0, 1]
print(f"\nCoefficiente di correlazione di Pearson: {pearson:.30f}")
````

**TASK3b: A partire dal dataset della Tabella di Mollier, implementare usando sia Excel che Python, graficando i dati (T vs. Vl) e calcolando i coeff. di Pearson e Spearman**

In questo task vengono analizzati i dati della Tabella di Mollier, che rappresenta le proprietà del vapore d'acqua tra temperatura (T) e volume specifico del vapore (Vi). Viene creato un grafico per visualizzare questa relazione e calcolati i coefficienti di correlazione di Pearson e Spearman, che aiutano a capire come sono legati in modo lineare questi 2 valori. Dal grafico si osserva una tendenza crescente tra temperatura e volume, ma con la presenza di un evidente outlier (in rosso) attorno ai 200 °C che distorce la relazione lineare. Il coefficiente di Pearson (0.3623) indica una correlazione lineare debole, mentre il coefficiente di Spearman (0.9592) evidenzia una forte correlazione monotona crescente, suggerendo che, pur non essendo perfettamente lineare, il volume tende ad aumentare in modo coerente con la temperatura.

image.png

````python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from scipy.stats import pearsonr, spearmanr
from sklearn.linear_model import LinearRegression

# Funzione per formattare i valori sull'asse delle y a 4 cifre decimali
def format_func(value, tick_number):
    return f'{value:.4f}'

# Definisce l'URL pubblico del file excel e il nome del foglio
public_url = f"https://docs.google.com/spreadsheets/d/{id_3}"
sheet_name = "tabella_mollier"

# Legge un file excel che viene scaricato tramite l'endpoint /export
df = pd.read_excel(public_url + "/export?format=xlsx", sheet_name=sheet_name)

# Estrazione dei valori di Temperatura e Volume da Excel
x = df["T (°C)"].values.reshape(-1, 1)
y = df["Vi (l)"].values

# Creazione del modello di regressione lineare
model = LinearRegression(fit_intercept=False)
model.fit(x, y) # Addestramento modello sui nostri dati
y_fit = model.predict(x) # Previsione valori

# Creazione del grafico a punti (scatter plot con outlier in rosso)
plt.figure(figsize=(8,5)) # Creazione grafico
outlier_index = df["Vi (l)"].idxmax() # Identificazione outlier (valore massimo di Vi (l))
plt.scatter(x=df["T (°C)"], y=df["Vi (l)"], color="yellow", edgecolor="black", s=100)
plt.scatter(df["T (°C)"].iloc[outlier_index], df["Vi (l)"].iloc[outlier_index], color="red", edgecolor="black", s=100)
plt.xlabel("Temperatura (°C)")
plt.ylabel("Volume (l)")
plt.title("Grafico Temperatura vs Volume")
plt.grid()
plt.xlim(0, 400)
plt.ylim(0, 0.25)
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(format_func)) # Mostra 4 cifre decimali sull'asse delle y
plt.plot(x, y_fit, color='black', linestyle='--') # Retta di regressione

# Mostra il grafico
plt.show()

# Calcolo dei coefficienti di correlazione e stampa
pearson_coef, _ = pearsonr(df['T (°C)'], df['Vi (l)'])
spearman_coef, _ = spearmanr(df['T (°C)'], df['Vi (l)'])
print(f"\nCoefficiente di Pearson: {pearson_coef:.4f}")
print(f"Coefficiente di Spearman: {spearman_coef:.4f}")
````

**TASK3c: Implementare es. del traffico milanese usando sia Excel che Python,
graficando i dati e calcolando il coeff. di correlazione (con shift)**

In questo task viene analizzato l’andamento del traffico milanese su base giornaliera per rilevare pattern ricorrenti nel tempo. Viene utilizzata l'autocorrelazione, una tecnica statistica che misura quanto una serie temporale sia correlata con sé stessa spostata nel tempo (shift). Nel codice vengono confrontati i dati della prima settimana con quelli delle settimane successive per vedere se ci sono regolarità settimanali. In base ai risultati mostrati nel grafico di autocorrelazione, si osservano picchi positivi regolari ogni 7 giorni, indicando la presenza di una certa stagionalità settimanale nel traffico milanese. Questo significa che il traffico tende a seguire uno schema simile di settimana in settimana, suggerendo abitudini ricorrenti nella mobilità urbana. Queste informazioni possono essere utili per ottimizzare la gestione del traffico e pianificare interventi nei giorni più critici.

image.png

````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Definisce l'URL pubblico del file excel e il nome del foglio
public_url = f"https://docs.google.com/spreadsheets/d/{id_3}"
sheet_name = "traffico_milanese"

# Legge un file excel che viene scaricato tramite l'endpoint /export
df = pd.read_excel(public_url + "/export?format=xlsx", sheet_name=sheet_name)

# Fissa i valori della prima settimana
fixed_wk = df['Traffico'][:7]

# Calcola l'autocorrelazione tra la prima settimana e le successive con shift
autocorrelations = []
for shift in range(1, len(df['Traffico']) - 6):
  lag = df['Traffico'][shift:shift + 7]
  autocorrelation = np.corrcoef(fixed_wk, lag)[0, 1]
  autocorrelations.append(autocorrelation)

# Crea i grafici con evidenza sui punti riscontrati
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Grafico Traffico
ax1.plot(range(len(df['Traffico'])), df['Traffico'], marker='o', linestyle='-', color='r', markersize=5)
ax1.set_title('Traffico', fontsize=18)
ax1.set_xlabel('Giorni', fontsize=12)
ax1.set_ylabel('Traffico', fontsize=12)
ax1.set_xticks(range(len(df['Traffico'])))
ax1.grid(True)

# Grafico Autocorrelazione
ax2.plot(range(1, len(autocorrelations) + 1), autocorrelations, marker='o', linestyle='-', color='b', markersize=5)
ax2.set_title('Autocorrelazione', fontsize=18)
ax2.set_xlabel('Shift in Giorni', fontsize=12)
ax2.set_ylabel('Coeff. Autocorrelazione', fontsize=12)
ax2.set_xticks(range(1, len(autocorrelations) + 1))
ax2.grid(True)

plt.tight_layout()
plt.show()
````

**Task5b: TrueSkill Ranking
Studiare TrueSkill e implementare in Python un esempio di calcolo automatico del punteggio TrueSkill. Creare qualche simulazione di calcolo del punteggio, anche per partite a squadre.**

Il ranking è un metodo per classificare i giocatori in base alla loro abilità, utile per determinare chi è più forte in un dato contesto. Un approccio semplice sarebbe dire che "chi ha più vittorie è il migliore", ma questo può non essere la scelta migliore: magari un giocatore ha affrontato solo avversari deboli. Per questo esistono sistemi come Elo e TrueSkill, che cercano di stimare la reale abilità di un giocatore.

*   Elo-ranking: Tutti i giocatori iniziano con lo stesso punteggio. Il sistema aggiorna il loro valore dopo ogni partita in base alla forza dell’avversario. Se vinci contro un avversario forte, il tuo punteggio aumenta di più rispetto a una vittoria contro un avversario debole.
*   TrueSkill: È un'evoluzione di Elo che tiene conto dell'incertezza nel ranking. Ogni giocatore ha un'abilità stimata con una "deviazione standard", che si riduce man mano che gioca. Questo permette di fare stime più precise, soprattutto quando i giocatori non si affrontano tutti tra loro.

Il codice presentato simula l’uso del sistema di ranking TrueSkill in due contesti: uno scontro 1 contro 1 e una partita a squadre. TrueSkill è un sistema sviluppato da Microsoft per stimare l'abilità dei giocatori in scenari competitivi, tenendo conto non solo del punteggio medio (mu) ma anche dell’incertezza (sigma) della stima.

Nel primo esempio, Alice e Bob iniziano con lo stesso rating (mu=25, sigma=8.333). Dopo la vittoria di Alice, il suo punteggio sale a mu=29.396 e l’incertezza scende, indicando maggiore fiducia nel fatto che sia più forte. Bob, al contrario, vede il suo rating scendere a mu=20.604, con una sigma ridotta.

Nel secondo esempio, si simula una partita 2v2 tra due squadre con rating iniziali uguali. Dopo la vittoria del Team 1, entrambi i giocatori vincenti aumentano il loro punteggio (mu=28.108), mentre quelli del Team 2 lo vedono calare (mu=21.892). In tutti i casi, anche la deviazione standard (sigma) si riduce, poiché il sistema guadagna fiducia nella stima dell’abilità.

Questa simulazione mostra come TrueSkill aggiorna dinamicamente i punteggi in base all'esito e alla qualità degli avversari, fornendo una valutazione più robusta rispetto al semplice conteggio delle vittorie.

````python
# Installazione libreria trueskill
!pip install trueskill
````

````python
import trueskill as ts

def simulate_1v1():
    # Inizializza i giocatori con il rating predefinito
    players = {
        "Alice": ts.Rating(),
        "Bob": ts.Rating()
    }

    print("== Partita 1vs1 ==")
    print("Rating iniziali:")
    for name, rating in players.items():
        print(f"  {name}: {rating}")

    # Simulazione della partita: Alice vince su Bob.
    team1 = [players["Alice"]]
    team2 = [players["Bob"]]
    new_ratings = ts.rate([team1, team2], ranks=[0, 1]) # 0 primo posto, 1 secondo posto

    # Aggiorna i rating dei giocatori con i nuovi valori
    players["Alice"] = new_ratings[0][0]
    players["Bob"] = new_ratings[1][0]

    print("\nDopo la partita (Alice vince su Bob):")
    for name, rating in players.items():
        print(f"  {name}: {rating}")

def simulate_team_match():
    # Inizializza i giocatori per la partita a squadre
    players = {
        "Team1_Player1": ts.Rating(),
        "Team1_Player2": ts.Rating(),
        "Team2_Player1": ts.Rating(),
        "Team2_Player2": ts.Rating()
    }

    print("\n== Partita a Squadre (Team vs Team) ==")
    print("Rating iniziali:")
    for name, rating in players.items():
        print(f"  {name}: {rating}")

    # Definisce le squadre:
    team1 = [players["Team1_Player1"], players["Team1_Player2"]]
    team2 = [players["Team2_Player1"], players["Team2_Player2"]]

    # Simula la partita: Squadra 1 (team1) vince su Squadra 2 (team2)
    new_ratings = ts.rate([team1, team2], ranks=[0, 1])

    # Aggiorna i rating dei giocatori con i nuovi valori
    players["Team1_Player1"] = new_ratings[0][0]
    players["Team1_Player2"] = new_ratings[0][1]
    players["Team2_Player1"] = new_ratings[1][0]
    players["Team2_Player2"] = new_ratings[1][1]

    print("\nDopo la partita (Squadra 1 vince su Squadra 2):")
    for name, rating in players.items():
        print(f"  {name}: {rating}")

def main():
    simulate_1v1()
    simulate_team_match()

if __name__ == "__main__":
    main()
````

**Task5c: Who’s Bigger?
Implementare in Python un sistema di scoring e ranking, analizzando Wikipedia e vari e multi-fattori, per valutare ‘’who’s bigger’’ tra: Romolo, Cesare, Dante Alighieri,Leonardo Da Vinci, Benito Mussolini, Umberto II Re d’Italia, Rita Levi Montalcini, Manzoni, Mario Draghi.**

Lo script genera due classifiche distinte per un insieme di figure italiane mostrando come il loro impatto digitale possa essere misurato in modo oggettivo. A partire dai nomi dei personaggi, viene interrogata l’API di Wikipedia per ricavare informazioni quali il PageRank della pagina, la lunghezza del testo, il numero di sezioni e le visualizzazioni mensili. Ciascuna di queste metriche viene poi standardizzata per renderle confrontabili su una stessa scala e, per la graduatoria “celebrity”, si introduce un decadimento esponenziale inverso rispetto all’anno di nascita: in questo modo si favoriscono le figure più recenti, coerentemente con l’idea di popolarità odierna. Infine, lo script ordina i personaggi in base al punteggio di visibilità digitale, che combina PageRank e visualizzazioni, e in base al punteggio di gravitas, definito dalla struttura e dalla lunghezza della voce, restituendo così due graduatorie che riflettono rispettivamente la popolarità e la ricchezza enciclopedica di ciascun protagonista.

````python
# Installazione libreria wikipedia-api
!pip install wikipedia-api
````

````python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from collections import defaultdict
from datetime import datetime
import wikipediaapi
import networkx as nx
import requests
import numpy as np
import math

# Ottiene la data attuale in formato YYYYMMDD
current_date = datetime.now().strftime("%Y%m%d")

# Recupera informazioni su pagina Wikipedia dato il nome. Ritorna un dizionario con titolo, lunghezza del testo, numero di link e sezioni.
def get_wikipedia_page_info(name):
    #istanza api wikipedia in italiano
    wiki_wiki = wikipediaapi.Wikipedia(
        user_agent='my_agent',
        language='it',
        extract_format=wikipediaapi.ExtractFormat.WIKI
    )
    page = wiki_wiki.page(name)
    # print(f"Name: {name}\nTitle: {page.title}\nLength: {len(page.text)}\nLinks: {len(page.links)}\nSections: {len(page.sections)}\n")

    if not page.exists(): #controlla se esiste pagina
        return None

    # Creazione del grafo aggiungendo nodi e archi
    G = nx.DiGraph()
    G.add_node(page.title)
    for link in page.links.keys():
        G.add_node(link)
        G.add_edge(page.title, link)

    pr = sum(nx.pagerank(G).values()) #calcolo del page rank

    return {  # Restituzione del dizionario con le metriche
        "page_rank": pr,
        "title": page.title,
        "length": len(page.text),
        "page_section": len(page.sections)
    }

# Recupera il numero di visualizzazioni mensili di una pagina Wikipedia.
def get_pageviews(title):
    url = f"https://wikimedia.org/api/rest_v1/metrics/pageviews/per-article/en.wikipedia/all-access/all-agents/{title}/monthly/00000101/{current_date}"
    data = requests.get(url, headers={"User-Agent": "my_agent"}).json() #riceve dati in formato json
    views = sum([item['views'] for item in data.get("items", [])]) #somma tutte views
    return views

# Applica un decadimento esponenziale basato sull'anno di nascita (da meno pesi personaggi molto antichi).
def apply_temporal_decay(birth_year, lambda_decay=0.0001):
    current_year = datetime.now().year
    time_diff = current_year - birth_year
    decay = math.exp(-lambda_decay * time_diff)
    return decay

# Normalizza i dati
def normalize(data):
    scaler = StandardScaler()
    standardized_data = scaler.fit_transform(data)
    return standardized_data

#Calcola il ranking delle persone in base a metriche prese da Wikipedia
def compute_final_ranking(people, birth_years):
    # Crea dizionari per salvare le metriche di ogni persona
    page_ranks = {}
    length_scores = {}
    links_scores = {}
    views_scores = {}
    decay_scores = {}
    sections_scores = {}

    for person in people:
        info = get_wikipedia_page_info(person)  # Ottiene i dati da Wikipedia
        if info:
            page_ranks[person] = info["page_rank"]
            length_scores[person] = info["length"]
            views_scores[person] = get_pageviews(info["title"])
            sections_scores[person] = info["page_section"]

            birth_year = birth_years.get(person, 0)
            decay_scores[person] = apply_temporal_decay(birth_year)

    page_rank_values = np.array(list(page_ranks.values())).reshape(-1, 1)
    views_values = np.array(list(views_scores.values())).reshape(-1, 1)
    length_values = np.array(list(length_scores.values())).reshape(-1, 1)
    sections_values = np.array(list(sections_scores.values())).reshape(-1, 1)

    # Normalizzazione
    page_rank_normalized = normalize(page_rank_values)
    views_normalized = normalize(views_values)
    length_normalized = normalize(length_values)
    sections_normalized = normalize(sections_values)

    final_celebrity = {}
    final_gravitas = {}

    for i, person in enumerate(people):
        # Per il punteggio "celebrity", somma page rank e views
        final_celebrity[person] = (page_rank_normalized[i][0] + views_normalized[i][0]) * decay_scores.get(person, 1)

        # Per il punteggio "gravitas", somma lunghezza e sezioni
        final_gravitas[person] = length_normalized[i][0] + sections_normalized[i][0]

    # Ordina i risultati in ordine decrescente
    ranked_celebrity = sorted(final_celebrity.items(), key=lambda x: x[1], reverse=True)
    ranked_gravitas = sorted(final_gravitas.items(), key=lambda x: x[1], reverse=True)

    return ranked_celebrity, ranked_gravitas


# Dizionario con persone e anni di nascita
people = {
    "Romolo": -771,
    "Giulio Cesare": -100,
    "Dante Alighieri": 1265,
    "Leonardo da Vinci": 1452,
    "Benito Mussolini": 1883,
    "Umberto II di Savoia": 1904,
    "Rita Levi-Montalcini": 1909,
    "Alessandro Manzoni": 1785,
    "Mario Draghi": 1947,
    "Cristiano Ronaldo": 1985
}

# Calcola il ranking
celebrity_ranking, gravitas_ranking = compute_final_ranking(people.keys(), people)

# Stampa la classifica "celebrity"
print("Classifica Celebrity:")
for rank, (name, score) in enumerate(celebrity_ranking, start=1):
    print(f"{rank}. {name} - Score: {score:.3f}")

# Stampa la classifica "gravitas"
print("\nClassifica Gravitas:")
for rank, (name, score) in enumerate(gravitas_ranking, start=1):
    print(f"{rank}. {name} - Score: {score:.3f}")
````

# **Significatività Statistica**

**Task7a.
Implementare in Python un esempio su dati a vs
piacimento dell’uso del T-test (va bene anche reimplementare es. QI M vs. F visto in Excel)**

Questo script consiste in un'analisi statistica dell'alcalinità della cenere tra due classi di vino (classe 1 e classe 2) utilizzando il dataset ```wine``` fornito da ```sklearn.datasets```. L'alcalinità della cenere è una misura chimica che si riferisce alla concentrazione di composti alcalini (come carbonati e idrossidi) nelle ceneri residue che derivano dalla combustione delle piante o dei materiali vegetali, come l'uva o altre sostanze organiche. L'alcalinità della cenere è un parametro importante in vari ambiti e può avere un impatto significativo sulla qualità del vino stesso. Nel codice vengono calcolate la media, la mediana, la deviazione standard, l'asimmetria e la curtosi per le classi 1 e 2. Inoltre, viene verificata la normalità dei dati attraverso il test di Shapiro-Wilk, che è una tecnica di Normality-testing per stabilire se le distribuzioni seguono un andamento normale. Successivamente, viene eseguito un t-test per confrontare le medie dei due gruppi e verificare se la differenza tra di esse è statisticamente significativa al 95% e al 99% di confidenza. I risultati sono presentati in una tabella che include le statistiche descrittive, il valore t, il p-value e l'esito del test di significatività.

In conclusione, l'analisi mostra se il secondo gruppo di vini ha un'alcalinità della cenere mediamente più alta del primo gruppo e questa differenza è significativa al 95% ma non al 99%.

````python
import pandas as pd
import numpy as np
import scipy.stats as stats
from sklearn.datasets import load_wine
from tabulate import tabulate

# Caricamento dataset
wine = load_wine()
df = pd.DataFrame(wine.data, columns=wine.feature_names)
df['target'] = wine.target #lista delle classi messi nella colonna target

# Creazione dei gruppi basati sul target
group1 = df[df['target'] == 1]['alcalinity_of_ash'].to_numpy()  # gruppo 1
group2 = df[df['target'] == 2]['alcalinity_of_ash'].to_numpy()  # gruppo 2

# Funzione che calcola media, mediana e deviazione standard
def calculate_statistics(x):
    return np.mean(x), np.median(x), np.std(x, ddof=1)

# Calcolo delle statistiche per il dataset del gruppo 1 e gruppo 2
mean_1, median_1, std_1 = calculate_statistics(group1)
mean_2, median_2, std_2 = calculate_statistics(group2)

# Numero di osservazioni per ogni gruppo
n_1, n_2 = len(group1), len(group2)

# Calcolo dell'asimmetria (skewness) per i gruppi
# bias=False applica una correzione statistica per ottenere una stima non distorta (unbiased)
skewness_1 = stats.skew(group1, bias=False)
skewness_2 = stats.skew(group2, bias=False)

# Calcolo della curtosi (kurtosis) per i gruppi
# bias=False applica una correzione statistica per ottenere una stima più affidabile su campioni piccoli.
kurtosis_1 = stats.kurtosis(group1, bias=False)
kurtosis_2 = stats.kurtosis(group2, bias=False)

# Test di normalità di Shapiro-Wilk per verificare se i dati seguono una distribuzione normale
shapiro_1 = stats.shapiro(group1)
shapiro_2 = stats.shapiro(group2)

# Interpretazione del test di normalità: se p-value > 0.05, i dati possono essere considerati normali
normality_1 = 'Si' if shapiro_1.pvalue > 0.05 else 'No'
normality_2 = 'Si' if shapiro_2.pvalue > 0.05 else 'No'

# Calcolo manuale del valore t per il test t di Student (test per la differenza tra le medie di due gruppi)
t_numerator = abs(mean_1 - mean_2)  # Differenza assoluta tra le medie
t_denominator = np.sqrt((std_1**2 / n_1) + (std_2**2 / n_2))  # Denominatore con errore standard combinato
t_value = t_numerator / t_denominator  # t value finale

# Calcolo dei gradi di libertà utilizzando la formula n1 + n2 - 2 (assunzione di varianze uguali)
df = n_1 + n_2 - 2

# Calcolo del p-value per il test t
p_value = 2 * (1 - stats.t.cdf(t_value, df))

# Soglie di significatività
sign_95 = 0.05
sign_99 = 0.01

# Verifica se il risultato è statisticamente significativo
significant_95 = p_value < sign_95  # Significativo al 95%
significant_99 = p_value < sign_99  # Significativo al 99%

# Creazione della tabella con i risultati
data = [
    ['Statistiche', 'Gruppo 1', 'Gruppo 2'],
    ['Media', f'{mean_1:.3f}', f'{mean_2:.3f}'],
    ['Mediana', f'{median_1:.3f}', f'{median_2:.3f}'],
    ['Dev.Std', f'{std_1:.3f}', f'{std_2:.3f}'],
    ['Asimmetria', f'{skewness_1:.1f}', f'{skewness_2:.1f}'],
    ['Distribuzione Normale', normality_1, normality_2],
    ['Curtosi', f'{kurtosis_1:.1f}', f'{kurtosis_2:.1f}'],
    ['t-value', f'{t_value:.5f}', f'{t_value:.5f}'],
    ['Gradi di libertà', f'{df}', f'{df}'],
    ['p-value', f'{p_value:.5f}', f'{p_value:.5f}'],
    ['Significativo al 95%', 'Sì' if significant_95 else 'No', 'Sì' if significant_95 else 'No'],
    ['Significativo al 99%', 'Sì' if significant_99 else 'No', 'Sì' if significant_99 else 'No']
]

print(f'\nConfronto: alcalinità della cenere tra le classi 1 e 2\n')
print(tabulate(data, headers='firstrow', tablefmt='fancy_grid'))
````

Successivamente, viene stimata la distribuzione cumulativa teorica che ci si aspetterebbe se i dati fossero normalmente distribuiti, basandosi sulla funzione di distribuzione normale (CDF) con i parametri di media e deviazione standard calcolati dai dati stessi. La differenza tra la distribuzione cumulativa osservata e quella teorica viene calcolata attraverso il KS-Test, che misura la massima distanza verticale tra le due distribuzioni. Il valore di questa distanza massima viene confrontato con un valore critico teorico per stabilire se la differenza osservata è statisticamente significativa, indicando una potenziale deviazione dalla normalità. Infine, i risultati di queste analisi vengono presentati in forma tabellare e grafica, visualizzando il confronto tra le distribuzioni attese e reali per entrambi i gruppi.

In conclusione, sulla base dei valori della distanza massima di Kolmogorov-Smirnov ottenuti e confrontati con il valore critico di riferimento, si determina se i punteggi di QI per il gruppo femminile e per il gruppo maschile possono essere considerati provenienti da una distribuzione normale.

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from itertools import takewhile
from tabulate import tabulate

# Definisce l'URL pubblico del file excel
public_url = f"https://docs.google.com/spreadsheets/d/{id_07}"
sheet_name = "Kolmogorov-Smirnov"

# Legge un file excel che viene scaricato tramite l'endpoint /export
df = pd.read_excel(public_url + "/export?format=xlsx", sheet_name=sheet_name, header=1)

# Legge le colonne di un dataframe, estrae i valori fino al primo NaN e li converte in array numpy
qi_f = np.array([value for value in takewhile(lambda x: not pd.isna(x), df["Dataset1: QI F"])])
qi_m = np.array([value for value in takewhile(lambda x: not pd.isna(x), df["Dataset2: QI M"])])

# Ordina in modo crescente i valori degli array
qi_f.sort()
qi_m.sort()

# Funzione che calcola media, mediana e deviazione standard
def calculate_statistics(dataset):
    return np.mean(dataset), np.median(dataset), np.std(dataset, ddof=1)

# Calcolo delle statistiche per il dataset delle femmine e dei maschi
mean_f, median_f, std_f = calculate_statistics(qi_f)
mean_m, median_m, std_m = calculate_statistics(qi_m)

# Funzione per calcolare la distribuzione cumulativa attesa
def expected_distribution(n):
    return (n - 1) / len(qi_f)

# Calcolo della distribuzione attesa cumulativa (percentuali)
cumulative = np.arange(1, 12)
expected_distribution_f = (cumulative - 1) / len(qi_f)
expected_distribution_m = (cumulative - 1) / len(qi_m)

# Calcolo della distribuzione reale (uso della distribuzione normale)
real_distribution_f = [norm.cdf(x, loc=mean_f, scale=std_f) for x in qi_f]
real_distribution_m = [norm.cdf(x, loc=mean_m, scale=std_m) for x in qi_m]

# Calcolo della distanza tra distribuzione attesa e distribuzione reale
distance_f = np.abs(np.array(expected_distribution_f) - np.array(real_distribution_f))
distance_m = np.abs(np.array(expected_distribution_m) - np.array(real_distribution_m))

# Distanza massima KS per i dati F e M
ks_f = np.max(distance_f)
ks_m = np.max(distance_m)

# Valore critico oracolo per n = 11 e alpha = 0.05 (vediamo dal valore della tabella KS)
# ks_critical = np.sqrt(-0.5 * np.log(alpha / 2) / n)
ks_critical = 0.39122

normal_f = 'Sì' if ks_f < ks_critical else 'No'
normal_m = 'Sì' if ks_m < ks_critical else 'No'

# Creazione della tabella con i risultati
data = [
    ['Statistiche', 'Femmine', 'Maschi'],
    ['Media', f'{mean_f:.3f}', f'{mean_m:.3f}'],
    ['Dev.Std', f'{std_f:.3f}', f'{std_m:.3f}'],
    ['Valore critico oracolo', f'{ks_critical:.5f}', f'{ks_critical:.5f}'],
    ['Distanza massima KS', f'{ks_f:.5f}', f'{ks_m:.5f}'],
    ['Distribuzione Normale', normal_f, normal_m]
]

# Stampa della tabella con i risultati in un formato leggibile
print(tabulate(data, headers='firstrow', tablefmt='fancy_grid'))

# Creazione della tabella con il cumulativo nella colonna verticale
data_cumulative_table = [
    cumulative,
    expected_distribution_f,
    real_distribution_f,
    distance_f,
    expected_distribution_m,
    real_distribution_m,
    distance_m
]

formatted_data = [
    [f"{x:.4f}" for x in col]  # Formatta ogni valore nella colonna con 4 decimali
    for col in data_cumulative_table
]

# Creazione dei titoli per le colonne
headers = ["Cumulativo", "Distribuzione Attesa F", "Distribuzione Reale F",
           "Distanza F", "Distribuzione Attesa M", "Distribuzione Reale M", "Distanza M"]

# Creazione e visualizzazione della tabella
data_for_table = list(zip(*formatted_data))
print(tabulate(data_for_table, headers=headers, tablefmt='fancy_grid'))

# Creazione del subplot con due grafici
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Grafico 1: Distribuzione Attesa vs Reale per le femmine (qi_f)
ax1.plot(qi_f, expected_distribution_f, label="Distribuzione Attesa F", color='blue', marker='o')
ax1.plot(qi_f, real_distribution_f, label="Distribuzione Reale F", color='red', marker='o')
ax1.set_xlabel('Cumulativo')
ax1.set_ylabel('Probabilità')
ax1.set_title('Distribuzione Attesa vs Reale per Femminile (QI F)')
ax1.legend()
ax1.grid(True)

# Grafico 2: Distribuzione Attesa vs Reale per i maschi (qi_m)
ax2.plot(qi_m, expected_distribution_m, label="Distribuzione Attesa M", color='blue', marker='o')
ax2.plot(qi_m, real_distribution_m, label="Distribuzione Reale M", color='red', marker='o')
ax2.set_xlabel('Cumulativo')
ax2.set_ylabel('Probabilità')
ax2.set_title('Distribuzione Attesa vs Reale per Maschile (QI M)')
ax2.legend()
ax2.grid(True)

plt.tight_layout()  # Migliora la disposizione
plt.show()
````

Viene quindi utilizzato un contatore per registrare quante volte ciascuna permutazione viene scelta. Alla fine, vengono tracciate le frequenze delle permutazioni in un grafico. **Task9b. Implementare in Python l’Esempio2 del foglio ‘Valutazione_Modelli’ contenuto nel file excel 09_Esempi_Excel_Costruire_Modelli.xlsx**

Questo script analizza i risultati di un test diagnostico per diverse soglie di dosaggio, calcolando e visualizzando la curva Receiver Operating Characteristic (ROC) e altre metriche di valutazione. Vengono estratte le osservazioni negative e positive per ciascun intervallo di dosaggio e calcolati i dati cumulativi di queste osservazioni. A partire da questi dati cumulativi, si determinano il Tasso di Falsi Positivi (FPR) e il Tasso di Veri Positivi (TPR) per ogni soglia.
Successivamente, viene calcolata l'Area Sotto la Curva (AUC) approssimandola come la somma delle aree di segmenti sottesi dalla curva ROC. Viene anche calcolato l'Indice di Youden per ogni soglia, che rappresenta la differenza tra TPR e FPR e fornisce una misura delle prestazioni complessive del test. Infine viene generato un grafico della curva ROC, che visualizza la relazione tra TPR e FPR al variare delle soglie. La curva ROC ottenuta (AUC = 0.894) suggerisce che il modello diagnostico ha una buona capacità di distinguere tra i casi positivi e negativi. La curva si discosta significativamente dalla linea diagonale che rappresenta un classificatore casuale, indicando una performance superiore.

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tabulate import tabulate
from itertools import takewhile
import warnings

# Imposta il filtro per ignorare messaggi di warning dovuti a estensioni non supportate durante la lettura del file excel
warnings.simplefilter('ignore')

# Definisce l'URL pubblico del file excel
public_url = f'https://docs.google.com/spreadsheets/d/{id_09}'
sheet_name = 'Valutazione_Modelli'

# Legge un file excel che viene scaricato tramite l'endpoint /export
df = pd.read_excel(public_url + '/export?format=xlsx', sheet_name=sheet_name, header=19)

# Legge le colonne di un dataframe, estrae i valori fino al primo NaN e li converte in array numpy
dosage_list = [value for value in takewhile(lambda x: x != 'tot.', df['Dosage'])]
neg_observed = np.array([value for value in takewhile(lambda x: not pd.isna(x), df['Negative'])])[:-1]
pos_observed = np.array([value for value in takewhile(lambda x: not pd.isna(x), df['Positive'])])[:-1]

# Dati: cumulati per ogni soglia
neg_cum = np.insert(np.cumsum(neg_observed), 0, 0)
pos_cum = np.insert(np.cumsum(pos_observed), 0, 0)

# Totali (l'ultima riga dei cumulati)
tot_neg = neg_cum[-1]
tot_pos = pos_cum[-1]

# Calcolo FPR e TPR
fpr = 1 - (neg_cum / tot_neg)
tpr = 1 - (pos_cum / tot_pos)

youden_index = [tpr[i] - fpr[i] for i in range(len(fpr))]

auc_segments = []
for i in range(len(fpr) - 1):  # Fino al penultimo elemento
    auc_segment = (fpr[i] - fpr[i + 1]) * tpr[i]  # Formula dell'AUC per il segmento
    auc_segments.append(auc_segment)

# Calcolo dell'AUC totale
total_auc = sum(auc_segments)
auc_segments_full = auc_segments + [0]

# Costruzione righe
data = [
    ['Dosage'] + [None] + dosage_list,
    ['Observed negative'] + [None] + neg_observed.tolist(),
    ['Observed positive'] + [None] + pos_observed.tolist(),
    ['Cumulative negative'] + neg_cum.tolist(),
    ['Cumulative positive'] + pos_cum.tolist(),
    ['FPR'] + fpr.tolist(),
    ['TPR'] + tpr.tolist(),
    ['AUC segment'] + auc_segments_full,
    ['Youden Index'] + youden_index
]

# Riorganizzazione dei dati
data = list(zip(*data))

# Intestazioni e righe
headers = data[0]
rows = data[1:]

floatfmt = ['.3f'] * len(headers)  # Applica 3 decimali a tutte le colonne tranne la prima
# Stampa tabella
print(tabulate(rows, headers=headers, tablefmt='fancy_grid', floatfmt=floatfmt) + "\n\n")

# Plot della curva ROC
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'Curva ROC (AUC = {total_auc:.3f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Classificatore Casuale')
plt.scatter(fpr, tpr, color='red', s=50, zorder=5)

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('Tasso di Falsi Positivi (FPR)')
plt.ylabel('Tasso di Veri Positivi (TPR)')
plt.title('Curva ROC del Modello di Classificazione')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
````

Questo script esegue una simulazione Monte Carlo per analizzare il valore potenziale di un portafoglio composto da cinque titoli (A, B, C, D, E). Successivamente, per ogni scenario, viene calcolato il valore totale del portafoglio come la somma dei valori dei singoli titoli e al termine delle simulazioni, vengono calcolate la media e la deviazione standard del valore totale del portafoglio risultante da tutti gli scenari simulati. Infine, vengono calcolate e stampate le probabilità stimate di perdere denaro (valore del portafoglio inferiore alla media) e di guadagnare più di 5000€ rispetto alla media, basandosi sui risultati delle simulazioni effettuate.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statistics
import math

# Definisce l'URL pubblico del file excel
public_url = f"https://docs.google.com/spreadsheets/d/{id_09}"
sheet_name = "MonteCarlo"

# Legge un file excel che viene scaricato tramite l'endpoint /export
df = pd.read_excel(public_url + "/export?format=xlsx", sheet_name=sheet_name, header=1)

# Estrae i valori dalla prima riga (valori massimi) e dalla seconda riga (valori minimi) dato il titolo
titles = ['A', 'B', 'C', 'D', 'E']
max_values = [float(df[lettera][0]) for lettera in titles]
min_values = [float(df[lettera][1]) for lettera in titles]

# Calcola il totale del portafoglio per valori massimi e minimi
total_portfolio_max = sum(max_values)  # Totale dei valori massimi
total_portfolio_min = sum(min_values)  # Totale dei valori minimi

# Parametri di simulazione
max_error_allowed = 0.05  # 5%
num_iterations = 39  # Numero di iterazioni Monte Carlo

print(f"Titoli: {titles}")
print(f"Valori massimi: {max_values}")
print(f"Valori minimi: {min_values}")
print(f"Totale portafoglio valori max: {total_portfolio_max}€")
print(f"Totale portafoglio valori min: {total_portfolio_min}€")
print(f"Errore massimo richiesto: {max_error_allowed * 100}%")
print(f"Numero di iterazioni: {num_iterations}")

# Funzione per generare un valore casuale tra min e max
def random_value(min_val, max_val):
    return np.random.random() * (max_val - min_val) + min_val

# Crea un DataFrame per memorizzare tutte le iterazioni
iterations_df = pd.DataFrame(columns=['iterazione', 'A', 'B', 'C', 'D', 'E', 'tot_portafoglio'])

# Genera i dati per ogni iterazione
for i in range(num_iterations):
    # Genera valori casuali per ogni titolo usando la formula equivalente a CASUALE()*(max-min)+min
    a_val = random_value(min_values[0], max_values[0])
    b_val = random_value(min_values[1], max_values[1])
    c_val = random_value(min_values[2], max_values[2])
    d_val = random_value(min_values[3], max_values[3])
    e_val = random_value(min_values[4], max_values[4])

    # Calcola il totale del portafoglio per questa iterazione
    tot_val = a_val + b_val + c_val + d_val + e_val

    # Aggiunge i dati al DataFrame
    iterations_df.loc[i] = [i+1, a_val, b_val, c_val, d_val, e_val, tot_val]

# Calcola la media e deviazione standard del portafoglio
portfolio_min_max_values = [total_portfolio_max, total_portfolio_min]
portfolio_mean = statistics.mean(portfolio_min_max_values)
portfolio_std_dev = statistics.pstdev(portfolio_min_max_values)

# Calcola l'errore assoluto secondo la formula "=MEDIA(I3:I4)*K3"
absolute_error = portfolio_mean * max_error_allowed
print(f"Media del portafoglio: {portfolio_mean:.2f}€")
print(f"Deviazione standard: {portfolio_std_dev:.2f}€")
print(f"Errore assoluto (media * errore_max): {absolute_error:.2f}€")

# Calcola l'errore assoluto secondo la formula classica di Monte Carlo
num_iterations_mc = ((portfolio_std_dev * 3) / absolute_error)**2
print(f"N iterazioni Monte Carlo: {num_iterations_mc:.2f}€")

# Calcola la probabilità di perdere soldi (valore del portafoglio inferiore alla media)
count_loss = (iterations_df["tot_portafoglio"].iloc[2:] < 55000).sum()
probability_loss = count_loss / num_iterations_mc
print(f"Probabilità di perdere soldi (rispetto alla media): {probability_loss*100:.2f}%")

# Calcola la probabilità di guadagnare più di 5000€ rispetto alla media
count_gain = (iterations_df["tot_portafoglio"].iloc[2:] >= 60000).sum()
probability_gain = count_gain / num_iterations_mc
print(f"Probabilità di guadagnare >5000€ (rispetto alla media): {probability_gain*100:.2f}%")
````

# **Regressione**

**Task10a. Disegnare uno studio completo (i.e., fit e validation) di regressione (lineare o non lineare) su un vostro dataset già studiato in task precedenti, usando le librerie offerte da Python**

Questo script in Python utilizza il dataset ```wine``` fornito dalla libreria ```sklearn.datasets```, che contiene misurazioni chimiche di diversi campioni di vino. Il codice inizia caricando il dataset e i dati vengono convertiti in un DataFrame per facilitarne l’analisi. Successivamente, viene calcolata la matrice di correlazione tra tutte le caratteristiche numeriche per identificare la coppia di variabili che presenta la correlazione lineare più alta in valore assoluto, escludendo i valori sulla diagonale (che rappresentano l’autocorrelazione).

Nel caso di questo dataset, le due variabili con la correlazione più alta sono ```total_phenols``` e ```flavanoids```: la prima indica la quantità complessiva di fenoli (sostenze naturali che contribuiscono al colore, al sapore e all’effetto antiossidante del vino), mentre la seconda misura i flavonoidi (responsabili della sensazione di secchezza).

Una volta trovata la coppia di variabili più correlate, il codice seleziona una come variabile indipendente (x) e l’altra come variabile dipendente (y), e divide i dati in set di addestramento e di test, usando una proporzione del 70% per l’addestramento e del 30% per il test, utile per addestrare e valutare un modello di regressione lineare.

Dopo aver calcolato l’accuratezza e l’errore quadratico medio (MSE), viene poi estratta e stampata l’equazione della retta di regressione, indicando il coefficiente angolare e l’intercetta. Infine, i dati originali e la linea di regressione ottenuta vengono visualizzati in un grafico scatter, mettendo in evidenza visivamente la relazione tra le due variabili selezionate.

Dato che viene scelto un seed fisso (42) per la suddivisione del dataset in set di addestramento e test, i risultati del modello rimangono invariati a ogni esecuzione. In particolare, l'accuratezza è pari a 74.28%, l’errore quadratico medio (MSE) è 0.24, mentre l’equazione della retta di regressione calcolata dal modello è $y = 1.34x - 1.03$.

Possiamo quindi concludere che un aumento di fenoli nel vino potrebbe implicare anche un aumento della sensazione di astringenza (secchezza), conferendo al vino una struttura più solida e una maggiore capacità di invecchiamento.

````python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.datasets import load_wine

# Caricamento del dataset del vino
wine = load_wine()
df = pd.DataFrame(wine.data, columns=wine.feature_names)

# Calcola la matrice di correlazione
correlation_matrix = df.corr()

# Trova la coppia di colonne con la correlazione più alta (escludendo la diagonale)
correlation_matrix = correlation_matrix.abs()  # Usato abs per evitare valori negativi nella correlazione
correlation_matrix.values[range(correlation_matrix.shape[0]), range(correlation_matrix.shape[1])] = 0  # Rimuovi la diagonale
max_corr = correlation_matrix.stack().idxmax()  # Trova l'indice della massima correlazione

# Selezioniamo le due colonne con la correlazione più alta
x = df[[max_corr[0]]]  # Variabile indipendente
y = df[max_corr[1]]    # Variabile dipendente

# Suddivisione in set di addestramento e di test -- 70% addestramento e 30% test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

# Crea e addestra il modello lineare sui dati di addestramento
model = LinearRegression()
model.fit(x_train, y_train)

# Valutazione sul dataset di test
y_pred = model.predict(x_test)

# Metriche del modello
accuracy = model.score(x_test, y_test)
print(f"Accuratezza: {accuracy * 100:.2f}%")
mse = mean_squared_error(y_test, y_pred)
print(f"MSE: {mse:.2f}")

# Recupera i coefficienti del modello
coef = model.coef_[0]
intercept = model.intercept_

# Crea e stampa l'equazione del modello
equation = f"y = {coef:.2f}x + {intercept:.2f}"
print(f"Equazione del modello: {equation}\n")

# Visualizza graficamente la regressione lineare
plt.scatter(x, y, color='blue', label='Dati')
plt.plot(x, model.predict(x), color='red', label='Regressione lineare')
plt.legend()
plt.grid(True)
plt.title(f"Regressione Lineare: {max_corr[0]} vs {max_corr[1]}")
plt.xlabel(max_corr[0])
plt.ylabel(max_corr[1])
plt.show()
````

Successivamente, il modello di regressione logistica viene allenato utilizzando la tecnica One-vs-Rest (OvR), che gestisce il problema di classificazione multiclasse, e viene calcolata l'accuratezza, indicando la percentuale di predizioni corrette.

````python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from matplotlib.colors import ListedColormap
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Carica il dataset Wine (solo 2 feature per visualizzazione)
wine = load_wine()
x = wine.data[:, [0, 6]]  # Alcohol e Flavanoids
y = wine.target
class_names = wine.target_names

# Standardizza le feature
scaler = StandardScaler()
x = scaler.fit_transform(x)

# Split del dataset in training e test set (70% training, 30% test)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

# Allena OvR con regressione logistica sul training set
base_model = LogisticRegression()
model = OneVsRestClassifier(base_model)
model.fit(x_train, y_train)

# Accuracy sul test set
accuracy = model.score(x_test, y_test)
print("Accuratezza sul test set: {:.2f}%".format(accuracy * 100))

# Equazioni dei classificatori
print("\nEquazioni dei classificatori OvR:")
for idx, clf in enumerate(model.estimators_):
    coef = clf.coef_[0]
    intercept = clf.intercept_[0]
    equation = "y_{} = {:.2f} * x0 + {:.2f} * x1 + {:.2f}".format(idx, coef[0], coef[1], intercept)
    print("Classe {}: {}".format(class_names[idx][-1], equation))

# Definisci una mappa di colori discreta per le classi
colors = ['#FF9999', '#99CC99', '#66B3FF']  # rosa, verde, azzurro
cmap = ListedColormap(colors)

# Plot dei confini di decisione
h = .02  # passo della griglia
x_min, x_max = x[:, 0].min() - .5, x[:, 0].max() + .5
y_min, y_max = x[:, 1].min() - .5, x[:, 1].max() + .5
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

z = model.predict(np.c_[xx.ravel(), yy.ravel()])
z = z.reshape(xx.shape)

plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, z, cmap=cmap, alpha=0.3)

# Plot dei dati reali con gli stessi colori
for i, color in enumerate(colors):
    plt.scatter(x_test[y_test == i, 0], x_test[y_test == i, 1],
                c=color, label=class_names[i], edgecolor='k', s=60)

print()
plt.xlabel('Alcohol (standardized)')
plt.ylabel('Flavonoids (standardized)')
plt.title('Confini di decisione - OvR Logistic Regression (Wine Dataset)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
print("\n\n")

# Calcola e visualizza la matrice di confusione con heatmap
y_pred = model.predict(x_test)
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names, linewidths=0.5, linecolor='white', square=True)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Matrice di Confusione - OvR Logistic Regression')
plt.tight_layout()
plt.show()
````
